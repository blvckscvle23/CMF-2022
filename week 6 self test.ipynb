```python
sym = 'AAPL'
AAPL_data = pd.DataFrame(data[sym]).dropna()
AAPL_data.tail() #extracting only apple data from the table
```

```python
window = 10
AAPL_data['min'] = data[sym].rolling(window=window).min()
AAPL_data['mean'] = data[sym].rolling(window=window).mean()
AAPL_data['std'] = data[sym].rolling(window=window).std()
AAPL_data['median'] = data[sym].rolling(window=window).median()
AAPL_data['max'] = data[sym].rolling(window=window).max()
AAPL_data['ewma'] = data[sym].ewm(halflife=0.5, min_periods=window).mean()
AAPL_data.dropna().head()#due to the window set as 10 days, the data starts from the next 10th day counting from the first day(4th March)
```

```python
ax = AAPL_data[['min', 'mean', 'max']].iloc[-250:].plot(figsize=(10, 6), style=['g--', 'r--', 'y--'], lw=0.8)
data[sym].iloc[-200:].plot(ax=ax, lw=2.0);#shows data from the last 250 days
```

```python
AAPL_data['SMA1'] = data[sym].rolling(window=20).mean()
AAPL_data['SMA2'] = data[sym].rolling(window=60).mean()
AAPL_data[[sym, 'SMA1', 'SMA2']].tail()#this is to setup simple rolling window
```

```python
AAPL_data[[sym, 'SMA1', 'SMA2']].plot(figsize=(10, 6))
```

```python
AAPL_data.dropna(inplace=True)
AAPL_data['positions'] = np.where(AAPL_data['SMA1'] > AAPL_data['SMA2'],1,-1)
ax = AAPL_data[[sym, 'SMA1', 'SMA2', 'positions']].plot(figsize=(10, 6),secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))#used for trend analysis
```

```python
Corr_data = data[['SPY', '^VIX']].dropna()
Corr_data.tail()#used to show the correlation between two stickers and dropna is to remove any NA
```

```python
Corr_data.plot(subplots=True, figsize=(10, 6))
```

```python
Corr_data.loc['2020-01-01':].plot(secondary_y='^VIX', figsize=(10, 6));
```

```python
sprets = np.log(Corr_data / Corr_data.shift(1))
sprets.head()
```

```python
sprets.dropna(inplace=True)
sprets.plot(subplots=True, figsize=(10, 6))
```

```python
pd.plotting.scatter_matrix(sprets, alpha=0.2, diagonal='hist', hist_kwds={'bins': 35}, figsize=(10, 6))#we want to find the linear correlation between SPY and VIX
```

```python
reg = np.polyfit(sprets['SPY'], sprets['^VIX'], deg=1)
ax = sprets.plot(kind='scatter', x='SPY', y='^VIX', figsize=(10, 6))
ax.plot(sprets['SPY'], np.polyval(reg, sprets['SPY']), 'r', lw=2);#we want to create a linear regression model for VIX and SPY
```

```python
sprets.corr()#correlation of two measures
```

```python
ax = sprets['SPY'].rolling(window=20).corr(sprets['^VIX']).plot(figsize=(10, 6)) 
ax.axhline(sprets.corr().iloc[0, 1], c='r');#red line is the correlation SPY/VIX of -0.83
```
